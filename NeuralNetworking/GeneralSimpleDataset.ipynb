```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
from network import Network
import torch

n_events = 1700
n_features = 16
```

```python
# hh[:,0] = np.random.rand(n_events)
# hh[:,1] = np.random.rand(n_events)
# qcd[:, 0] = -1 * np.random.rand(n_events)
# qcd[:, 1] = -1 * np.random.rand(n_events)

hh = np.random.rand(n_events, n_features)
qcd = np.random.rand(n_events, n_features)


hh = np.abs(hh)
qcd = np.abs(qcd)

hh = hh * (1/np.max(hh))
qcd = -qcd * (1/np.max(qcd))

# add labels
hh_labels= np.zeros((len(hh),1))
hh_labels = hh_labels+1
hh = np.append(hh, hh_labels, axis=1)

qcd_labels= np.zeros((len(qcd),1))
qcd = np.append(qcd, qcd_labels, axis=1)


print(hh[:4])
print(qcd[:4])

```

```python
all_data = np.append(hh, qcd, axis=0)
print(all_data[:4])
for i in range(4):
    np.random.shuffle(all_data)
print(all_data[:4])
```
